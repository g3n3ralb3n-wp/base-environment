## Definition of List ##

**list** (*noun*) [Python Programming]

A list is a fundamental, built-in data structure in Python that represents an ordered, mutable sequence of elements enclosed in square brackets `[]`, capable of storing zero or more items of any data type.

**Key Characteristics:**

1. **Ordered**: Maintains insertion order with each element at a specific index position (zero-based indexing).

2. **Mutable**: Contents can be modified after creation—elements may be added, removed, or replaced.

3. **Heterogeneous**: Can contain elements of different data types (integers, strings, floats, objects, or other lists).

4. **Indexed**: Elements are accessible via zero-based integer indexing, supporting both positive (0, 1, 2...) and negative indices (-1, -2...).

5. **Dynamic**: Automatically expands or contracts as elements are added or removed.

6. **Iterable**: Supports iteration through loops and comprehensions.

**Syntax:** `[element1, element2, element3, ...]`

**Common Operations:** Access (`list[index]`), modification (`list[index] = value`), append (`list.append(item)`), insert, remove, and slicing.

```python
# Import libraries
import pandas as pd
from sklearn.datasets import load_iris

# Import iris dataset
iris = load_iris()

# Create DataFrame
iris_df = pd.DataFrame(
    data=iris.data,
    columns=iris.feature_names
)
iris_df['target'] = iris.target
```

```python
# Create a list using the first observation (all features and label)
first_observation = iris_df.iloc[0].tolist()
iris_list = first_observation

# Show output with print statement
print("List with first observation:")
print(iris_list)
print()

# Add the second observation to the list
second_observation = iris_df.iloc[1].tolist()
iris_list.extend(second_observation)

# Show output with print statement
print("List after adding second observation:")
print(iris_list)
```

# (b) Dictionary
- Define a "dictionary" as comprehensively and accurately as you would find in a dictionary. Provide a clear and concise explanation, covering its essential meaning, characteristics, and any relevant context.
- Create a dictionary in Python using the classes of Setosa, Versicolor and Virginica, allow elements (2 observations per class) to be added
- Show the output with a print statement
- Add a third observation of the Iris data set for each class
- Show the output of your with a print statement

## Definition of Dictionary ##

**dictionary** (*noun*) [Python Programming]

A dictionary is a built-in, mutable data structure in Python that stores an unordered collection of key-value pairs enclosed in curly braces `{}`, where each unique key maps to a corresponding value.

**Key Characteristics:**

1. **Key-Value Pairs**: Stores data as associated pairs where keys act as unique identifiers for their corresponding values.

2. **Unordered**: Prior to Python 3.7, dictionaries had no guaranteed order. From Python 3.7+, insertion order is preserved as an implementation detail (officially guaranteed in Python 3.8+).

3. **Mutable**: Contents can be modified after creation—key-value pairs can be added, removed, or updated.

4. **Keys Must Be Unique**: Each key can appear only once; duplicate keys will overwrite previous values.

5. **Keys Must Be Immutable**: Keys must be of an immutable type (strings, numbers, tuples), while values can be any data type.

6. **Fast Lookup**: Uses hash tables internally, providing O(1) average time complexity for access, insertion, and deletion operations.

**Syntax:** `{key1: value1, key2: value2, key3: value3, ...}`

**Common Operations:** Access (`dict[key]`), assignment (`dict[key] = value`), deletion (`del dict[key]`), iteration through keys/values/items.

```python
# Create a dictionary with iris classes (2 observations per class)
# Setosa: target = 0, Versicolor: target = 1, Virginica: target = 2

# Get observations for each class
setosa_obs = iris_df[iris_df['target'] == 0].iloc[:2]
versicolor_obs = iris_df[iris_df['target'] == 1].iloc[:2]
virginica_obs = iris_df[iris_df['target'] == 2].iloc[:2]

# Create dictionary with class names as keys
iris_dict = {
    'Setosa': setosa_obs.values.tolist(),
    'Versicolor': versicolor_obs.values.tolist(),
    'Virginica': virginica_obs.values.tolist()
}

# Show output with print statement
print("Dictionary with 2 observations per class:")
for class_name, observations in iris_dict.items():
    print(f"{class_name}: {observations}")
print()

# Add a third observation for each class
setosa_third = iris_df[iris_df['target'] == 0].iloc[2].tolist()
versicolor_third = iris_df[iris_df['target'] == 1].iloc[2].tolist()
virginica_third = iris_df[iris_df['target'] == 2].iloc[2].tolist()

iris_dict['Setosa'].append(setosa_third)
iris_dict['Versicolor'].append(versicolor_third)
iris_dict['Virginica'].append(virginica_third)

# Show output with print statement
print("Dictionary after adding third observation for each class:")
for class_name, observations in iris_dict.items():
    print(f"{class_name}: {observations}")
```

# (c) Tuple
- Define a "tuple"" as comprehensively and accurately as you would find in a dictionary. Provide a clear and concise explanation, covering its essential meaning, characteristics, and any relevant context.
- Create a tuple in Python using the classes of Setosa, Versicolor and Virginica, allow elements ( start with 2 observations and 2 features per class) to be added
- Show the output with a print statement
- Add a third observation and a third feature of the Iris data set for each class
- Show the output of your with a print statement

## Definition of Tuple ##

**tuple** (*noun*) [Python Programming]

A tuple is a built-in, immutable data structure in Python that represents an ordered sequence of elements enclosed in parentheses `()`, capable of storing zero or more items of any data type.

**Key Characteristics:**

1. **Ordered**: Maintains a fixed sequence with each element at a specific index position (zero-based indexing).

2. **Immutable**: Once created, elements cannot be modified, added, or removed. The tuple's contents are permanently fixed.

3. **Heterogeneous**: Can contain elements of different data types (integers, strings, floats, objects, lists, or other tuples).

4. **Indexed**: Elements are accessible via zero-based integer indexing, supporting both positive (0, 1, 2...) and negative indices (-1, -2...).

5. **Hashable**: Due to immutability, tuples can be used as dictionary keys or elements in sets (provided all elements are also hashable).

6. **Memory Efficient**: Generally consume less memory than lists and have faster iteration due to immutability.

**Syntax:** `(element1, element2, element3, ...)` or `element1, element2, element3` (parentheses optional but recommended for clarity)

**Common Operations:** Access (`tuple[index]`), slicing, concatenation, unpacking, and iteration. Note: Modification operations like append, remove, or item assignment are not supported.

```python
# Create tuples with 2 observations and 2 features per class
# Note: Since tuples are immutable, we'll use tuple concatenation to "add" elements

# Get 2 observations with first 2 features for each class
setosa_data = iris_df[iris_df['target'] == 0].iloc[:2, :2]
versicolor_data = iris_df[iris_df['target'] == 1].iloc[:2, :2]
virginica_data = iris_df[iris_df['target'] == 2].iloc[:2, :2]

# Create tuple with nested tuples for each class
iris_tuple = (
    ('Setosa', tuple(tuple(row) for row in setosa_data.values)),
    ('Versicolor', tuple(tuple(row) for row in versicolor_data.values)),
    ('Virginica', tuple(tuple(row) for row in virginica_data.values))
)

# Show output with print statement
print("Tuple with 2 observations and 2 features per class:")
for class_name, observations in iris_tuple:
    print(f"{class_name}: {observations}")
print()

# Add a third observation and third feature for each class
# Since tuples are immutable, we create a new tuple by concatenation
setosa_third = tuple(iris_df[iris_df['target'] == 0].iloc[2, :3])
versicolor_third = tuple(iris_df[iris_df['target'] == 1].iloc[2, :3])
virginica_third = tuple(iris_df[iris_df['target'] == 2].iloc[2, :3])

# Create new tuple with additional observations and features
iris_tuple_updated = (
    ('Setosa', tuple(tuple(iris_df[iris_df['target'] == 0].iloc[i, :3]) for i in range(3))),
    ('Versicolor', tuple(tuple(iris_df[iris_df['target'] == 1].iloc[i, :3]) for i in range(3))),
    ('Virginica', tuple(tuple(iris_df[iris_df['target'] == 2].iloc[i, :3]) for i in range(3)))
)

# Show output with print statement
print("Tuple after adding third observation and third feature for each class:")
for class_name, observations in iris_tuple_updated:
    print(f"{class_name}: {observations}")
```

- Define a "set"" as comprehensively and accurately as you would find in a dictionary. Provide a clear and concise explanation, covering its essential meaning, characteristics, and any relevant context.
- Create a set (1st set) in Python using the first observation of the Setosa class allowing elements to be added
- Show the output with a print statement
- Create another set (2nd set) by adding the second observation of the Setosa class 
- Show the output of the following operations with the 1st set and 2nd set with a print statement
    * Difference
    * Intersection
    * Union

## Definition of Set ##

**set** (*noun*) [Python Programming]

A set is a built-in, mutable data structure in Python that represents an unordered collection of unique elements enclosed in curly braces `{}`, designed for mathematical set operations and membership testing.

**Key Characteristics:**

1. **Unordered**: Elements have no defined order or index. The order of elements may vary and is not guaranteed.

2. **Unique Elements**: Automatically eliminates duplicate values; each element can appear only once in the set.

3. **Mutable**: The set itself can be modified after creation—elements can be added or removed (though elements must be immutable types).

4. **Unhashable Elements Not Allowed**: Elements must be of immutable types (integers, strings, tuples). Lists, dictionaries, or other sets cannot be set elements.

5. **No Indexing**: Since sets are unordered, elements cannot be accessed by index or position.

6. **Fast Membership Testing**: Uses hash tables internally, providing O(1) average time complexity for checking if an element exists.

7. **Mathematical Operations**: Supports set operations like union, intersection, difference, and symmetric difference.

**Syntax:** `{element1, element2, element3, ...}` or `set()` for empty set (note: `{}` creates an empty dictionary)

**Common Operations:** Add (`set.add(item)`), remove (`set.remove(item)`), union (`set1 | set2`), intersection (`set1 & set2`), difference (`set1 - set2`), membership testing (`item in set`).

```python
# Create 1st set using first observation of Setosa class
# Note: Sets can only contain immutable types, so we'll convert to tuples
setosa_first = iris_df[iris_df['target'] == 0].iloc[0]

# Create set from first observation (converting to tuple since lists are not hashable)
set_1 = set([tuple(setosa_first.values)])

# Show output with print statement
print("1st set with first observation of Setosa class:")
print(set_1)
print()

# Create 2nd set by adding second observation of Setosa class
setosa_second = iris_df[iris_df['target'] == 0].iloc[1]
set_2 = set([tuple(setosa_second.values)])

# Show output with print statement
print("2nd set with second observation of Setosa class:")
print(set_2)
print()

# Perform set operations
print("Set Operations:")
print("-" * 50)

# Difference (elements in set_1 but not in set_2)
difference = set_1 - set_2
print(f"Difference (set_1 - set_2): {difference}")
print()

# Intersection (elements common to both sets)
intersection = set_1 & set_2
print(f"Intersection (set_1 & set_2): {intersection}")
print()

# Union (all unique elements from both sets)
union = set_1 | set_2
print(f"Union (set_1 | set_2): {union}")
```
